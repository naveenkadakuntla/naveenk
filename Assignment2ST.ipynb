```python
""" importing the modules """
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def readCSV(l):
    """ reading the csv file """
    df = pd.read_csv(l)
    c = df.iloc[:,0:1]
    y = df.iloc[:,1:]  
    return c,y

def stat_data(cl,y):
    """ Computing the statistics of the data """
    s = y.apply(pd.DataFrame.describe, axis=1)
    s.insert(0, 'Countries', cl)
    print('\nStatistical Data')
    display(s)
    print("\n\n")

def linePlot(cl,y):
    """ plotting the line graph for the data """
    yl = list(y.columns)
    yd = y.values.tolist()
    plt.figure(figsize=[10,10])
    for i in range(8):
        plt.plot(yl,yd[i],lw = 3,label = cl[i])
    plt.legend()
    plt.xlabel('Year')
    plt.ylabel('Cereal production (metric tons)')
    plt.title("Cereal production in different countries", size = 20)
    plt.show()
    plt.close()

l = "/content/cereal_production.csv"
c,y = readCSV(l)
cl = c['Country Name'].tolist()
display(c,y)

stat_data(cl,y)

linePlot(cl,y)

data = pd.read_csv("/content/japan.csv")

""" Mapping the correlations on different indicators """
cor = sns.heatmap(data.corr(), annot=True,cmap='RdYlGn',linewidths=0.30)
plt.title("Correlation on different indicators for Japan\n\n")
display(cor)
```
